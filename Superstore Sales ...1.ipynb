```python
def get_Sales_level(Sales):
    if Sales > 10000:
        return 'costly'
    elif Sales > 1000:
        return 'medium'
    else :
        return 'cheap'
```

```python
months_df1 = df1.groupby("Month").sum().sort_values("Sales", ascending = False)
months_df1 = months_df1[["Sales"]]
months_df1.reset_index(inplace=True)
months_df1
```

```python
plt.bar(months_df1["Month"],months_df1["Sales"], color = "pink", edgecolor = "red" ,linewidth = .5)
plt.xlabel("Month", size = 20)
plt.ylabel("Sales", size = 20)
plt.xticks(fontsize = 10, rotation = 70)
plt.yticks(fontsize = 10)
plt.title("Sales based on Month", size = 20, pad = 20)
plt.savefig('Sales based on Month.jpg')
plt.show()
```

```python
MonthSales_data = df1.groupby(by=['Year','Month']).sum()
Total_Sales_per_Year = MonthSales_data.reset_index()
Total_Sales_per_Year['Month'] = Total_Sales_per_Year.Month.apply(lambda x:x[:3])
MonthSales_data =Total_Sales_per_Year.groupby(by=['Year', 'Month']).sum()['Sales']
fig, ax = plt.subplots(nrows=4, ncols=1, figsize=(10, 8))

yrs = [2015, 2016, 2017, 2018]
for i in range(4):
    yr = yrs[i]
    Total_Sales_per_Year = MonthSales_data.loc[yr]
    ax[i] = sns.lineplot(x= Total_Sales_per_Year.index, y=Total_Sales_per_Year.values, data=Total_Sales_per_Year, 
                         ax=ax[i], label=yr, marker="o", color="#3F007190")
    ax[i].set_ylabel('Sales')

plt.show()
```

```python
Total_Sales_per_Year = pd.DataFrame(df1.groupby(by=['Year']).sum())

plt.figure(figsize=(14, 4))
sns.pointplot(x=Total_Sales_per_Year.index, y='Sales', data=Total_Sales_per_Year)

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title("Total Sales per Year")

plt.show()
```
